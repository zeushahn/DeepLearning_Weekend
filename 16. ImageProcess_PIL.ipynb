#### Thumbnail 만들기
- 가로와 세로의 크기를 반으로 줄인다.
- 반으로 줄이면 실수(Real Number)가 발생할 수 있어 소수점 이하 truncate

```python
import math
```

```python
math.trunc(cat.size[0]/2)
```

```python
# Thumbnail 만들기
size = math.trunc(cat.size[0]/2), (math.trunc(cat.size[1]/2))
catThumb = cat.copy()
catThumb.thumbnail(size)
catThumb.save("../Data/thumb.png")
catThumb
```

#### Image Merge

```python
record = Image.open("../Data/record.png")
record
```

```python
# mode 확인
record.mode
```

```python
# 크기 확인
record.size
```

```python
# 이미지의 크기가 cat보다 커서 줄이자
# 1/10 크기를 줄여서 recordThumb를 만들자
size = math.trunc(record.width / 10), math.trunc(record.height/10)
recordThumb = record.copy()
recordThumb.thumbnail(size)
recordThumb.save("../Data/record_thumb.png")
recordThumb
```

```python
# cat Image(169, 158)위에 recordThumb를 배치
position = (50, 50)
catMerge = cat.copy()
catMerge.paste(recordThumb, position)
catMerge
```

> recordThumb의 배경색 투명(검은색)이고 이미지도 검은색    
검은색 투명도 제거 필요 

```python
# cat Image(169, 158)위에 recordThumb를 배치
position = (50, 50)
catMerge = cat.copy()
catMerge.paste(recordThumb, position, recordThumb)
catMerge
```

#### Center에 배치

```python
# cat Image(169, 158)위에 recordThumb를 배치
position = (math.trunc((cat.width - recordThumb.width)/2)), \
            (math.trunc((cat.height - recordThumb.height)/2))
catMerge = cat.copy()
catMerge.paste(recordThumb, position, recordThumb)
catMerge
```
